```python
import pandas as pd
```

```python
_id = [1, 2, 3, 4, 5, 6]
balance = [5229.12, 12727.52, 14996.58, 21149.0, 14379.0, 6221.12]
interest_rate = [8.75, 11.37, 8.25, 4.75, 3.75, 6.75]
rate_type = ['variable', 'fixed', 'fixed', 'variable', 'variable', 'variable']
loan_id = [2, 4, 9, 7, 5, 11]
df=pd.DataFrame({'id':_id,'balance':balance,'interest_rate':interest_rate,'rate_type':rate_type,'loan_id':loan_id})
df
```

```python

fixed=[]
varibale=[]
for i in range(len(df)):
    if df['rate_type'][i]=='fixed':
        fixed.append(1)
        varibale.append(0)
    else:
        fixed.append(0)
        varibale.append(1)
res=pd.DataFrame({'loan_id':df['loan_id'],'fixed':fixed,'varibale':varibale})
res
```
