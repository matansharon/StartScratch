```python
res={}
for i in range(len(df)):
    row = df.iloc[i]
    day=row.order_timestamp_utc.strftime('%Y-%m-%d')
    if day not in res:
        res[day] = {}
    if row.order_timestamp_utc.hour not in res[day]:
        res[day][row.order_timestamp_utc.hour] = 0
    res[day][row.order_timestamp_utc.hour] += 1
m=-1
hour=[]
for date in res:
    if max(res[date].values())>m:
        m=max(res[date].values())
for date in res:
    for h in res[date]:
        if res[date][h]==m and h not in hour:
            hour.append(h)
n_orders=[m]*len(hour)
df1=pd.DataFrame({'hour':hour,'n_orders':n_orders}).sort_values(by='hour',ascending=True)
df1
            
```
