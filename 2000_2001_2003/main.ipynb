```python
total_fixed=df['balance'][df['rate_type']=='fixed'].sum()
total_variable=df['balance'][df['rate_type']=='variable'].sum()
balance_share=[]
for i in range(len(df)):
    if df['rate_type'][i]=='fixed':
        balance_share.append((df['balance'][i]/total_fixed*100))
    else:
        balance_share.append(df['balance'][i]/total_variable*100)
df['balance_share']=balance_share
res=pd.DataFrame({'loan_id':df['loan_id'],'rate_type':df['rate_type'],'balance':df['balance'],'balance_share':df['balance_share']})
res
```
