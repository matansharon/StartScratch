```python
import pandas as pd
year =[2019, 2019, 2020, 2019, 2019, 2019, 2020, 2019, 2020, 2020, 2019, 2020, 2019, 2020, 2019, 2020, 2020, 2020, 2019, 2019]
company_name =['Toyota', 'Toyota', 'Toyota', 'Honda', 'Honda', 'Honda', 'Honda', 'Honda', 'Chevrolet', 'Chevrolet', 'Chevrolet', 'Chevrolet', 'Ford', 'Ford', 'Ford', 'Jeep', 'Jeep', 'Jeep', 'Jeep', 'Jeep']
product_name =['Avalon', 'Camry', 'Corolla', 'Accord', 'Passport', 'CR-V', 'Pilot', 'Civic', 'Trailblazer', 'Trax', 'Traverse', 'Blazer', 'Figo', 'Aspire', 'Endeavour', 'Wrangler', 'Cherokee', 'Compass', 'Renegade', 'Gladiator']
#create a dataframe with the following columns as year, company_name, product_name
df = pd.DataFrame({'year':year, 'company_name':company_name, 'product_name':product_name})
```

```python
res={}
for i in range(len(df)):
    row=df.iloc[i]
    if row.company_name not in res:
        res[row.company_name]={2019:0, 2020:0}
    res[row.company_name][row.year]+=1
names=res.keys()
diff=[]
for i in res:
    diff.append(res[i][2020]-res[i][2019])
df2=pd.DataFrame({'company_name':names, 'net_new_products':diff})
df2.sort_values(by='company_name', inplace=True)
df2
```
