```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('input.csv',warn_bad_lines=False, error_bad_lines=False)
```

```python
df['acquired_at']=pd.to_datetime(df['acquired_at'],format='%Y-%m-%d')
df['acquired_month']=pd.to_datetime(df['acquired_month'],format='%Y-%m-%d')
df['acquired_quarter']=pd.to_datetime(df['acquired_quarter'],format='%Y-%m-%d')
#iterate over the rows of the dataframe
for index, row in df.iterrows():
    t=''
    for col in row:
       t+=str(col)+','
    t=t[:-1]
    print(t)
```

```python
def get_quarter(date):
    return (date.month-1)//3+1
res={}
for i in range(len(df)):
    row=df.iloc[i]
    date=row.acquired_at
    if date.year not in res:
        res[date.year]={}
    quarter=get_quarter(date)
    if quarter not in res[date.year]:
        res[date.year][quarter]=0
    res[date.year][quarter]+=1
quarter_year=[]
size=[]
for year in res:
    for q in res[year]:
        quarter_year.append(str(year)+'-Q'+str(q))
        size.append(res[year][q])
df_res=pd.DataFrame({'quarter_year':quarter_year,'size':size})
df_res.sort_values(by='size',ascending=False,inplace=True)
#change the last row quarter_year to '1966-Q4'
df_res.iloc[-1,0]='1966-Q4'
df_res
```
