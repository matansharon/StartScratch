```python
pathes={}
for i in range(len(test_df)):
    row=test_df.iloc[i]
    if row.origin not in pathes:
        pathes[row.origin]={}
    if row.destination not in pathes[row.origin]:
        pathes[row.origin][row.destination]=row.cost


pathes
```

```python
res={}

for i in pathes:
    
    for j in pathes[i]:
        # print(i,j,pathes[i][j])
        if (i,j) not in res:
            res[(i,j)]=pathes[i][j]
        if j in pathes:
           for k in pathes[j]:
            #    print(i,k,pathes[i][j]+pathes[j][k])
                if (i,k) not in res:
                    res[(i,k)]=pathes[i][j]+pathes[j][k]
                else:
                    res[(i,k)]=min(res[(i,k)],pathes[i][j]+pathes[j][k])
res
        


        
```

```python
origin=[]
destinations=[]
costs=[]
for i in res:
    origin.append(i[0])
    destinations.append(i[1])
    costs.append(res[i])
# print(origin)
# print(destinations)
# print(costs)
res_df=pd.DataFrame({'origin':origin,'destination':destinations,'cost':costs})
res_df.sort_values(by=['origin','destination','cost'],inplace=True)
res_df
```
