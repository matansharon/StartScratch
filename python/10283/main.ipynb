```python
import pandas as pd
from datetime import datetime
```

```python
df=pd.read_csv('data.csv')
```

```python
top=[]
for i in range(len(df)):
    row=df.iloc[i]
    if row.year_rank==1 and row.year>=datetime.now().year-20:
        top.append(row.song_name)
top=set(top)
top=pd.DataFrame(top,columns=['song_name'])
top
```
