```python
import pandas as pd
import numpy as np
import os 
from os import listdir
from random import randint,choice,random
from datetime import datetime, timedelta
```

```python
#get the path of the current file
path=os.getcwd()

data_path=path+'/data/'
data_path
group_names=[]
with open(data_path+'names.txt','r') as f:
    for line in f:
        group_names.append(line[:-1])

artists=[]
with open(data_path+'artists.txt','r') as f:
    for line in f:
        artists.append(line[:-1])
songs=[]
with open(data_path+'songs.txt','r') as f:
    for line in f:
        songs.append(line[:-1])
year=range(1965,2016)
year_rank=range(1,100)
ids=range(1,10000)
df=pd.DataFrame({
    'id':np.random.choice(ids,10000),
    'year':np.random.choice(year,10000),
    'year_rank':np.random.choice(year_rank,10000),
    'group_name':np.random.choice(group_names,10000),
    'artist':np.random.choice(artists,10000),
    'song':np.random.choice(songs,10000)
}   
)


```

```python
year_2010=df[df['year']==2010]
res=year_2010.sort_values(by='year_rank',ascending=True)
#drop duplicates
res=res.drop_duplicates(subset='group_name',keep='first')
res=res.head(10)
#get only the year_rank, group_name, artist, song columns
res=res[['year_rank','group_name','artist','song']]
res
```
