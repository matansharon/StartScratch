```python
class Customer:
    def __init__(self,customer_id,):
        self.customer_id=customer_id
        self.stores={}
    def add(self,store_brand,sale):
        if store_brand not in self.stores:
            self.stores[store_brand]=[]
        self.stores[store_brand].append(sale)
    def toString(self):
        return str(self.customer_id)+str(self.stores)
```

```python
class Store:
    def __init__(self,store_name):
        self.store_name=store_name
        self.high={'customers':set(),'sales':[]}
        self.medium={'customers':set(),'sales':[]}
        self.low={'customers':set(),'sales':[]}
    def add(self,customer_id,sale):
        if sale>=30:
            self.high['customers'].add(customer_id)
            self.high['sales'].append(sale)
        elif sale<30 and sale>=20:
            self.medium['customers'].add(customer_id)
            self.medium['sales'].append(sale)
        else:
            self.low['customers'].add(customer_id)
            self.low['sales'].append(sale)
    def toString(self):
        res=str(self.store_name)+'\n'
        res+=str(len(self.high['customers']))+' '+str(len(self.high['sales']))+' '+str(sum(self.high['sales'])/len(self.high['sales']))+'\n'
        res+=str(len(self.medium['customers']))+' '+str(len(self.medium['sales']))+' '+str(sum(self.medium['sales'])/len(self.high['sales']))+'\n'
        res+=str(len(self.low['customers']))+' '+str(len(self.low['sales']))+' '+str(sum(self.low['sales'])/len(self.high['sales']))+'\n'
        return res
```
