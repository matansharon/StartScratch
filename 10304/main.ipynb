```python
class Project:
    def __init__(self,_id,title,budget,start,end):
        self.id=_id
        self.title=title
        self.budget=budget
        s=start.split('-')
        e=end.split('-')
        start=datetime(int(s[0]),int(s[1]),int(s[2]))
        end=datetime(int(e[0]),int(e[1]),int(e[2]))
        self.duration=(end-start).days
        self.emps=set()
        self.prorated_expence=0
    def add_emp(self,emp_id,salary):
        if emp_id in self.emps:
            return
        self.emps.add(emp_id)
        self.prorated_expence+=salary*(self.duration/365)
        
```

```python
res={}
for i in range(len(df4)):
    row=df4.iloc[i]
    if row['project_id'] not in res:
        res[row['project_id']]=Project(row['project_id'],row['title'],row['budget'],row['start_date'],row['end_date'])
    res[row['project_id']].add_emp(row['emp_id'],row['salary'])
```

```python
res_title=[]
res_budget=[]
res_prorated_expense=[]
for i in res:
    curr=res[i]
    if curr.prorated_expence>curr.budget:
        res_title.append(curr.title)
        res_budget.append(curr.budget)
        res_prorated_expense.append(ceil(curr.prorated_expence))
df5=pd.DataFrame({'title':res_title,'budget':res_budget,'prorated_expense':res_prorated_expense})
df5
```
